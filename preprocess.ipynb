```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import word_tokenize
from nltk.tokenize import TweetTokenizer
import pandas as pd
import re
import contractions
```

```python
raw_comments_df=pd.read_csv('raw_reddit_data_filtered.csv')
raw_comments=raw_comments_df['Comment']
```

```python
comm_lines = []
for comm in raw_comments:
    comm_lines.extend(comm.split("_x000D_"))

comm_lines_filterd = []
regex = r'\S'
for string in comm_lines:
    if re.search(regex, string):
        comm_lines_filterd.append(string)

comm_sentences = []
for comm in comm_lines:
    comm_sentences.extend(sent_tokenize(comm))
```

```python
comm_sentences_expanded = []
for comm in comm_sentences:
    comm_sentences_expanded.append(contractions.fix(comm))

comm_sentences_expanded
```

```python
comm_sent_symb = []
for comm in comm_sentences_expanded:
    lst = TweetTokenizer().tokenize(comm)
    lst_syb= []
    reg = r'[^0-9a-zA-Z]'
    for string in lst:
        if not re.match(reg, string):
            if not (re.match(r'https://',string) or re.match(r'http://',string)):
                lst_syb.append(string)
        # elif len(string)>1:
        #     lst_syb.append(string)
    comm_sent_symb.append(" ".join(lst_syb))

comm_sent_symb
```

```python
comm_sent_symb_filterd = []
regex = r'\S'
for string in comm_sent_symb:
    if re.search(regex, string):
        comm_sent_symb_filterd.append(string)
```

```python
comm_preprocessed = pd.Series(comm_sent_symb_filterd)
comm_preprocessed = comm_preprocessed.sample(frac=1,random_state=0)
train_set = comm_preprocessed.iloc[:int(0.8*len(comm_preprocessed))].reset_index()
test_set = comm_preprocessed.iloc[int(0.8*len(comm_preprocessed)):].reset_index()
train_set.to_csv("train_set_preprocessed_1.csv",index=False)
test_set.to_csv("test_set_preprocessed_1.csv",index=False)
```
